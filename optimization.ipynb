```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -0.01553 * x**5 - 0.09659 * x**4 + 0.1198 * x**3 + 1.148 * x**2 + 0.149 * x - 1.019
```

```python
f = np.poly1d([-0.01553, -0.09659, 0.1198, 1.148, 0.149, -1.019])
```

```python
xs = np.linspace(-5, 3, 100)
ys = f(xs)
plt.plot(xs, ys)
plt.plot(-0.1, f(-0.1), "o")
```
